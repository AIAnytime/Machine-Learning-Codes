```python
def custom_accuracy(y_test,y_pred):
    right = 0

    l = len(y_pred)
    for i in range(0,l):
        if(abs(y_pred[i]-y_test[i]) <= 20):
            right += 1
    print((right/l)*100)
```

```python
X = mydata.iloc[:,[8,9,10,11,12,13,14,15,16]].values
y = mydata.iloc[:, 17].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin = LinearRegression()
lin.fit(X_train,y_train)
```

```python
y_pred = lin.predict(X_test)
print(lin.score(X_test,y_test)*100)
```

```python
import numpy as np
new_prediction = lin.predict(sc.transform(np.array([[101,1,19,5.32,27,0,5.4,71,12]])))
print(new_prediction)
```
